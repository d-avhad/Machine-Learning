### (b) Train a classifier for each label

#### (i) Research

<font color='blue'>
1. Exact match : <br>
In multilabel multioutput classification, exact match sees whether all the labels are exactly matching the predicted labels. If yes then a score of 1 is given, else a score of 0 is given. <br> <br>
2. Hamming loss : <br> 
In multilabel classification, hamming loss is calculated as the avg of the hamming distance between the predicted and true value. instead of a strict 1 or 0 policy, it counts number of mismatches between each label (this is same as hamming distance between 2 vectors). the average of this mismatch count is hamming loss. <br>
Hamming score: The measure of the number of labels that are correctly predicted out of all the labels present in the dataset. The score is calculated as the average of the number of labels correctly predicted for each instance in the dataset.

```python
# function to calculate hamming loss and exact match

def cal_metrics(true_y, pred_y, single=False):
    
    # calculate hamming loss 
    
    miss_labels = 0 
    for t,p in zip(true_y.values, pred_y.values):
        miss_count = (t != p)
        miss_labels += np.sum(miss_count)
    
    if(single==True):
        hamming_loss = miss_labels / (true_y.shape[0])
    else:
        hamming_loss = miss_labels / (true_y.shape[0] * true_y.shape[1])
    
    
    # calculate exact match ratio
    
        
    exact_match = 0
    for t, p in zip(true_y.values, pred_y.values):
        count_match = (t == p)
        
        if(single==True):
            if(count_match==True):
                exact_match += 1
        else:        
            if sum(count_match) == true_y.shape[1]:
                exact_match += 1;
                
    exact_match /= true_y.shape[0]
    
    hamming_loss = np.round(hamming_loss, 4)
    exact_match = np.round(exact_match, 4)
    
    return hamming_loss, exact_match
```

#### (ii) Train a SVM for each of the labels

```python

hamming_loss=dict()
exact_matc=dict()
best_c=dict()
best_gamma=dict()
print('For Gaussian SVM :')

#training and finding best params for each label

for l in labels:
    
    svc=SVC(kernel='rbf',decision_function_shape='ovr')
    
    paramgrid={'C':[10**i for i in range(-3,3)],'gamma':[i/10 for i in range(1,20)]}
    
    stratified_cv=StratifiedKFold(n_splits=10)
    
    clf=GridSearchCV(svc,param_grid=paramgrid,cv=stratified_cv)
    
    clf.fit(train_x, train_y[l])
    
    print('Label=',l,'Best score=',clf.best_score_,'Best params',clf.best_params_)
    
    best_c[l]=clf.best_params_['C']
    best_gamma[l]=clf.best_params_['gamma']
    
```

```python
#fiting model with best params  for each label
y_pred=dict()

for l in labels:
    s=SVC(C=best_c[l],gamma=best_gamma[l],kernel="rbf",decision_function_shape='ovr').fit(train_x,train_y[l])
    y_pred[l] = s.predict(test_x)
    
```

```python
predictY = pd.DataFrame(columns=test_y.columns)
for label in test_y.columns:
    
    test_pred = y_pred[label]
    predictY.loc[:, label] = test_pred
    
    
print("(Hamming Score, Hamming Loss, Exact Match Ratio) for Label Individually:")

for l in labels:
    print(l)
   
    h,e=cal_metrics(test_y[l],predictY[l],True)
    print((1-h),h,e)
```

```python
print("(Hamming Score, Hamming Loss, Exact Match Ratio) FOR ALL LABELS:")
    

h_l,e_m=cal_metrics(test_y,predictY)
print("Hamming Score",1-h_l)
print("Hamming Loss",h_l)
print("Exact match",e_m)
```

#### (iii) Repeat 1(b)ii with L1-penalized SVMs

```python
# Here I am using pipeline for standardizing and then training. Pipeline does everything regarding standardizing data,
#for example  the following is done internally:
# The data are split into TRAINING data and TEST data according to the cv parameter  specified in the GridSearchCV.
# the scaler is fitted on the TRAINING data
# the scaler transforms TRAINING data

y_pred_l1=dict()

best_c_l1=dict()

p=[{'linear__C':[10**i for i in range(-3,3)]}]

print('------L1 PENALIZED SVM:---------')

for l in labels:
    
    svc_linear_clf=LinearSVC(penalty='l1', loss = 'squared_hinge',dual=False,max_iter=1000000)
    
    pipe_line = Pipeline([
    ('scale', StandardScaler()),
    ('linear', svc_linear_clf)])
    
    st_cv=StratifiedKFold(n_splits=10)
    
    clf=GridSearchCV(pipe_line,param_grid=p,cv=st_cv)
    
    clf.fit(train_x,train_y[l])   # training the model using the classifier with best parameters.
    
    print('Label=',l,'Best score=',clf.best_score_,'Best params',clf.best_params_)
    
    best_c_l1[l]=clf.best_params_['linear__C']
    
    y_pred_l1[l]=clf.predict(test_x) # predicting the model using the classifier with best parameters.
        



```

```python
predictY = pd.DataFrame(columns=test_y.columns)
for label in test_y.columns:
    
    test_pred = y_pred_l1[label]
    predictY.loc[:, label] = test_pred
    
print("FOR SVM WITH L1 PENALTY")    
print("(Hamming Score, Hamming Loss, Exact Match Ratio) for Label Individually:")

for l in labels:
    print(l)
   
    h,e=cal_metrics(test_y[l],predictY[l],True)
    print((1-h),h,e)
```

## 2. K-Means Clustering on a Multi-Class and Multi-Label Data Set

### (a) Use k-means clustering

Monte Carlo Simulation: 
In following code I have used total 50 iterations, under each iteration running k-means for k=1,2...50 and finding best k in each iteration.

```python
# function to get the optimal value of K

def calculate_optimal_k(num, X, rand):
    
    best_k = 2
    max_val=0
    num = num + 1 
    
    for n in range(2, num):
        cl = KMeans(n_clusters=n)
        cluster_labels = cl.fit_predict(X)
        
        sil_val = silhouette_score(X, cluster_labels)
        
        if sil_val > max_val:
            best_k = n
            max_val = sil_val
            
    print(f"\nOPTIMAL K : {best_k}")
    print("\nSILHOUTTE_SCORE: ",max_val)
    
    return best_k
```

```python
#  function to get the majority labels of a cluster

def calculate_majority(bestk, cl_lab, Y):
    
    major_cl = pd.DataFrame(columns=Y.columns)
    
    for c in range(bestk):
        
        idx, = np.where(cl_lab == c)
        cluster_samples = Y.iloc[idx, :]
        row = []
        for l in Y.columns:
            cur_major = cluster_samples.loc[:, l].value_counts().index[0]
            row.append(cur_major)
            
        major_cl.loc[c] = row
        
    return major_cl
```

```python
# function to calculate  hamming distance and hamming loss for majority

def calucalate_hamming(major_cl, cl_lab, y_val):
    
    missclf_labels = 0
    
    for c in range(len(major_cl)):
        
        idx, = np.where(cl_lab == c)
        
        for label in y_val.loc[idx].values:
            miss = (label != major_cl.loc[c].values)
            missclf_labels += np.sum(miss)
       
    
    hamming_dist = missclf_labels / y_val.shape[0]
    hamming_loss = missclf_labels / (y_val.shape[0] * y_val.shape[1])
    
    return hamming_dist, hamming_loss
```

```python
def monte_carlo_func(itr, X, Y):
    
    hamming_dist = []
    hamming_loss = []
    
    for i in range(itr):
        
        print("\n-------------------FOR ITERATION",i + 1,"---------------------------------------")
        
        #get best k
        bestk = calculate_optimal_k(50, X, i)
        
        # get clusterer using best k
        clt = KMeans(n_clusters=bestk)
        
        #fit clusterer on data
        cluster_labels = clt.fit_predict(X)
        
        # get majority labels in each cluster
        cluster_major = calculate_majority(bestk, cluster_labels, Y)
        
        print("\nMAJORITY LABELS FOR EACH CLUSTERS (INDEX=CLUSTER)")
        print(cluster_major)
        
        # calculate the hamming distance and hamming loss (avg of all clusters in a iteration)
        cur_dist, cur_loss = calucalate_hamming(cluster_major, cluster_labels, Y)
        hamming_dist.append(cur_dist)
        hamming_loss.append(cur_loss)
        print(f" AVG HAMMING DISTANCE: {round(cur_dist, 6)}, AVG HAMMING LOSS: {round(cur_loss, 6)}")
        print("-----------------------------------------------------------------------------------")
        
    return hamming_dist, hamming_loss
```

```python
#50 iterations
# Monte-Carlo Simulation on whole dataset
hamming_dist, hamming_loss = monte_carlo_func(50, mfcc_data.iloc[:, :-4], mfcc_data.iloc[:, -4:-1])

```

### (c) Calculate the average Hamming distance, Hamming score, and Hamming loss

```python
def summary(hamming_distance, hamming_loss):
    summary_hd = {
        "Average Hamming Distance": [round(np.mean(hamming_distance), 6)],
        "Std Devation (Hamming Distance)": [round(np.std(hamming_distance), 6)]
    }
    
    summary_hl = {
        "Average Hamming Loss": [round(np.mean(hamming_loss), 6)],
        "Std Devation (Hamming Loss)": [round(np.std(hamming_loss), 6)]
    }
    
    summary_h_score = {
        "Average Hamming Score": [round(1 - np.mean(hamming_loss), 6)],
        "Std Devation (Hamming Score)": [round(np.std(hamming_loss), 6)]
    }
    
    df_summary_hd = pd.DataFrame(data=summary_hd)
    df_summary_hl = pd.DataFrame(data=summary_hl)
    df_summary_h_score = pd.DataFrame(data=summary_h_score)
    
    return [df_summary_hd, df_summary_hl, df_summary_h_score]
```

```python
df_hamming_distance, df_hamming_loss, df_hamming_score = summary(hamming_dist, hamming_loss)
```
