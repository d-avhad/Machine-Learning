### (c) Convert the decision rules

```python
def tree_to_decision_rules(tree, feature_names):
    left = tree.tree_.children_left
    right = tree.tree_.children_right
    threshold = tree.tree_.threshold
    features = [feature_names[i] for i in tree.tree_.feature]
    value = tree.tree_.value

    def recurse(left, right, threshold, features, node, depth=0):
        indent = "  " * depth
        if (threshold[node] != -2):
            print(indent,"if ( " + features[node] + " <= " + str(threshold[node]) + " ) {")
            if left[node] != -1:
                recurse (left, right, threshold, features, left[node], depth+1)
                print(indent,"} else {")
                if right[node] != -1:
                    recurse (left, right, threshold, features, right[node], depth+1)
            print(indent,"}")
        else:
            print(indent,"return " + str(value[node]))

    recurse(left, right, threshold, features, 0)
```

```python
tree_to_decision_rules(diagnosis_model,['Temperature(C)','Nausea','Lumbar pain','Urine pushing',
                                            'Micturition pains','Burning/Itching/Swelling of Bladder'])
```

### (d) Calculate the Coefficient of Variation CV

```python
class CoefficientVariation():
    def __init__(self,data_set,fnames):
        self.fnames = fnames
        self.features = {}
        self.data = data_set
        self.mean = self.generate_means()
        self.std_dev = self.generate_std_deviation()
        self.coeff_variation = self.generate_coeff_variation()
        
    def generate_means(self):
        means = []
        for i in range(0,123):
            column = list(self.data.iloc[:,i])
            means.append(self.get_sum(column)/len(column))
        return means
    
    def get_sum(self,column):
        csum = 0
        for num in column:
            csum += num
        return csum
    
    def generate_std_deviation(self):
        std_dev = []
        for i in range(0,123):
            column = list(self.data.iloc[:,i])
            std_dev.append((self.get_sum(self.get_vsquare(column,self.mean[i]))/len(column))**0.5)
        return std_dev
    
    def get_vsquare(self,col,mean):
        vsquare = []
        for num in col:
            vsquare.append((num-mean)**2)
        return vsquare
    
    def generate_coeff_variation(self):
        cv = []
        for i in range(0,123):
            cv.append(self.std_dev[i]/self.mean[i])
            self.features[cv[i]] = self.fnames[i]
        return cv
            
    def get_coefficient_variation(self):
        return self.coeff_variation
    
    def get_feature_names(self):
        return self.features
```

### (g) Fit a ridge regression model

```python
import warnings
warnings.filterwarnings('ignore')
def build_models(model,x_train,y_train,x_test,y_test):
    m = model.fit(x_train,y_train)
    m_score = m.score(x_test,y_test)
    y_predict = m.predict(x_test)
    test_error = 1 - m_score
    m_mse = mean_squared_error(y_test,y_predict)
    print(f'Best lambda: {m.alpha_}')
    print(f'model score: {m_score}')
    print(f'model error: {test_error}')
    print(f'model mse: {m_mse}')
    return (m.alpha_,m_score,test_error,m_mse)
```

```python
lmbdas=[0.001, 0.01, 0.1, 1]
```

```python
rmodel = RidgeCV(alphas=lmbdas, cv=5, normalize= False)
```

```python
build_models(rmodel,xc_train,yc_train,xc_test,yc_test)
```

### (i) Fit a PCR model

```python
def get_best_mserror(bind,bmse,m_list,xc_train,yc_train):
    mserror = []
    for ind in m_list:
        p = PCA(n_components=ind)
        x_temp = p.fit_transform(xc_train)
        lreg = LinearRegression()
        cvs = cross_validate(lreg, x_temp, yc_train, cv=3,scoring=('r2', 'neg_mean_squared_error'))
        rmean = np.mean(cvs['test_r2'])
        mserror = -1*(np.mean(cvs['test_neg_mean_squared_error']))
        if mserror < bmse:
            bmse = mserror
            bind = ind
    return (bmse,bind)
```

```python
berr,bm = get_best_mserror(math.inf,math.inf,list(range(1,123)),xc_train,yc_train)
print(f'(best principle train components,best train mean square error) = {(berr,bm)}')
```

```python
M = bm
pca = PCA(n_components=M)
pca = pca.fit(xc_train)
p_train = pca.transform(xc_train)
p_test = pca.transform(xc_test)
linear = LinearRegression()
linear = linear.fit(p_train,yc_train)
lin_predict = linear.predict(p_test)
p_mse = mean_squared_error(yc_test,lin_predict)
print(f'PCA Test MSE: {p_mse}')
```

### (j) Fit a boosting tree

```python
alpha = []
msed = []
xgb_matrix = xgboost.DMatrix(data=xc_train, label=yc_train)
for lmbda in lmbdas:
    param = {'max_depth':3, 'alpha':lmbda}
    xgb_sc = xgboost.cv(param, xgb_matrix, num_boost_round=100, nfold=5,metrics='rmse')
    mse = np.mean(xgb_sc['test-rmse-mean'])
    alpha.append(lmbda)
    msed.append(mse)
```

```python
fin_pd = pd.concat([pd.DataFrame(alpha,columns=['Lambda']),pd.DataFrame(msed,columns=['Mean Squared Error'])],axis=1)
```
